```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')
tip2 = ctrl.Consequent(np.arange(0, 26, 1), 'tip2')

# Auto-membership function population is possible with .automf(3, 5, or 7)
quality.automf(3)
service.automf(3)

# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['low'] = fuzz.trimf(tip2.universe, [0, 0, 13]) #graph plotting
tip['medium'] = fuzz.trimf(tip2.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip2.universe, [13, 25, 25])

tip2['low'] = fuzz.trimf(tip2.universe, [0, 0, 13]) #graph plotting
tip2['medium'] = fuzz.trimf(tip2.universe, [0, 13, 25])
tip2['high'] = fuzz.trimf(tip2.universe, [13, 25, 25])
```

```python
quality['average'].view()
```

```python
service.view()
```

```python
tip.view()
```

```python
rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])
rule4 = ctrl.Rule(quality['poor'] | service['poor'], tip2['low'])
rule5 = ctrl.Rule(service['average'], tip2['medium'])
rule6 = ctrl.Rule(service['good'] | quality['good'], tip2['high'])

rule1.view()
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
```

```python
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```

```python
tipping.input['quality'] = 6.5
tipping.input['service'] = 9.8

# Crunch the numbers
tipping.compute()
```

```python
tipping.
```

```python
t=tipping.output['tip']
# t2=tipping.ouput['tip2']
tip.view(sim=tipping)
```

```python
import pickle
filename = 'fuzzy.sav'
pickle.dump(tipping, open(filename, 'wb'))
```

```python
loaded_model3 = pickle.load(open("fuzzy.sav","rb"))
loaded_model3.input['quality']=6.5
loaded_model3.input['service']=9.8
loaded_model3.compute()
t=loaded_model3.output['tip']
tip.view(sim=loaded_model3)
```
